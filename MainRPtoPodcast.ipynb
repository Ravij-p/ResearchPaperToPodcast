```python
# Install required libraries
!pip install PyPDF2 paddleocr gtts moviepy pillow python-pptx llama-cpp-python pdf2image pyngrok

# Import libraries
import os
import PyPDF2
import paddleocr
from gtts import gTTS
from PIL import Image, ImageDraw, ImageFont
from moviepy.editor import concatenate_audioclips, AudioFileClip
from llama_cpp import Llama
from pdf2image import convert_from_path
from collections import defaultdict
from pptx import Presentation
from pptx.util import Inches
from concurrent.futures import ThreadPoolExecutor, as_completed
from flask import Flask, render_template, request, send_file
from pyngrok import ngrok
import time

# Mount Google Drive (for accessing the LLaMA model)
from google.colab import drive
drive.mount('/content/drive')

# Initialize LLaMA model
model_path = "/content/drive/MyDrive/llama-3.2-3b-instruct-q8_0.gguf"  # Update this path
llm = Llama(
    model_path=model_path,
    n_ctx=1024,  # Reduced context window for faster processing
    n_threads=4  # Use multiple threads for inference
)

# Enhanced configuration with multiple voice profiles per role
ROLES = {
    "Scientist": {
        "variants": [
            {"tld": "com", "speed": False, "gender": "female"},  # Female voice (US English)
            {"tld": "co.uk", "speed": False, "gender": "male"}   # Male voice (UK English)
        ]
    },
    "Historian": {
        "variants": [
            {"tld": "com.au", "speed": True, "gender": "female"},  # Female (Australian)
            {"tld": "co.uk", "speed": False, "gender": "male"}     # Male (British)
        ]
    },
    "Philosopher": {
        "variants": [
            {"tld": "fr", "speed": False, "gender": "male"},      # Male (French accent)
            {"tld": "com", "speed": True, "gender": "female"}     # Female (US slow)
        ]
    }
}

# Backend functions (same as before)
def extract_text(pdf_path):
    """Hybrid text extraction for research papers with parallel OCR"""
    text = ""
    try:
        # Extract text using PyPDF2
        with open(pdf_path, "rb") as f:
            reader = PyPDF2.PdfReader(f)
            for page in reader.pages:
                text += page.extract_text() or ""

        # Use OCR if text extraction is insufficient
        if len(text) < 1000:
            images = convert_from_path(pdf_path, thread_count=4)  # Parallelize PDF to image conversion
            ocr = paddleocr.PaddleOCR(use_angle_cls=True, lang='en', use_gpu=True)  # Enable GPU for OCR

            def ocr_page(img):
                result = ocr.ocr(img)
                return "\n".join([line[-1][0] for line in result[0]])

            with ThreadPoolExecutor() as executor:
                futures = [executor.submit(ocr_page, img) for img in images]
                for future in as_completed(futures):
                    text += future.result() + "\n"

    except Exception as e:
        print(f"Extraction error: {str(e)}")
    return text

def generate_summary(text, llm):
    """Generate paper summary using LLM with optimized prompt"""
    prompt = f"Summarize this research paper in 3 key points (be concise):\n{text[:3000]}"
    response = llm(prompt, max_tokens=256)  # Reduced max_tokens for faster generation
    return response['choices'][0]['text']

def generate_dialogue(text, selected_roles, llm):
    """Generate interactive dialogue between participants with optimized prompt"""
    role_counts = defaultdict(int)
    participants = []
    for role in selected_roles:
        role_counts[role] += 1
        participants.append(f"{role} {role_counts[role]}" if role_counts[role] > 1 else role)

    prompt = f"""Generate a concise conversation between {len(participants)} academics discussing this research paper:
    Paper content: {text[:2000]}

    Participants: {", ".join(participants)}
    Requirements:
    - Alternate speakers naturally
    - Include questions and responses
    - Show different perspectives
    - 4-6 exchanges total
    - Format as "[Speaker]: [Message]" exactly"""

    response = llm(prompt, max_tokens=512)  # Reduced max_tokens for faster generation
    return response['choices'][0]['text']

def create_audio(script, selected_roles):
    """Generate multi-voice audio with parallel TTS"""
    clips = []
    temp_files = []
    role_instances = defaultdict(int)

    def generate_speech(speaker, content, variant):
        tts = gTTS(text=content.strip(), lang='en', tld=variant["tld"], slow=variant["speed"])
        temp_file = f"temp_{speaker.replace(' ', '_')}.mp3"
        tts.save(temp_file)
        return temp_file, AudioFileClip(temp_file)

    with ThreadPoolExecutor() as executor:
        futures = []
        for line in script.split('\n'):
            if ':' in line:
                speaker, content = line.split(':', 1)
                speaker = speaker.strip()
                base_role = speaker.split(' ')[0]

                # Get voice configuration
                role_config = ROLES.get(base_role, {"variants": [{"tld": "com", "speed": False}]})
                instance_num = role_instances[base_role]
                variant = role_config["variants"][instance_num % len(role_config["variants"])]

                # Submit TTS task to thread pool
                futures.append(executor.submit(generate_speech, speaker, content, variant))
                role_instances[base_role] += 1

        for future in as_completed(futures):
            temp_file, clip = future.result()
            temp_files.append(temp_file)
            clips.append(clip)

    # Combine audio clips
    if clips:
        final_audio = concatenate_audioclips(clips)
        final_audio.write_audiofile("dialogue.mp3")

    # Cleanup
    for f in temp_files:
        try:
            os.remove(f)
        except:
            pass

def create_summary_visual(summary):
    """Generate summary visualization"""
    img = Image.new("RGB", (1920, 1080), "white")
    draw = ImageDraw.Draw(img)

    try:
        font = ImageFont.truetype("arial.ttf", 24)
    except:
        font = ImageFont.load_default()

    y = 50
    for line in summary.split('\n'):
        draw.text((50, y), line, font=font, fill="black")
        y += 30

    img.save("summary_visual.png")

def create_ppt(summary, dialogue):
    """Generate PowerPoint presentation with minimal formatting"""
    prs = Presentation()
    slide_layout = prs.slide_layouts[1]

    # Add summary slide
    slide = prs.slides.add_slide(slide_layout)
    title = slide.shapes.title
    content = slide.shapes.placeholders[1].text_frame
    title.text = "Research Paper Summary"
    content.text = summary

    # Add dialogue slide
    slide = prs.slides.add_slide(slide_layout)
    title = slide.shapes.title
    content = slide.shapes.placeholders[1].text_frame
    title.text = "Dialogue"
    content.text = dialogue

    prs.save("research_paper_presentation.pptx")

# Flask app
app = Flask(__name__)

# Create the uploads directory if it doesn't exist
os.makedirs("uploads", exist_ok=True)

# Start Ngrok and get the public URL
try:
    public_url = ngrok.connect(5000).public_url
    print(f" * Ngrok URL: {public_url}")
except Exception as e:
    print(f"Error starting Ngrok: {e}")

@app.route('/')
def index():
    return render_template('index.html')

@app.route('/upload', methods=['POST'])
def upload():
    if 'file' not in request.files:
        return "No file part"

    file = request.files['file']
    selected_roles = request.form.getlist('roles')

    if file.filename == '':
        return "No selected file"

    if file:
        file_path = os.path.join("uploads", file.filename)
        file.save(file_path)

        # Process the file
        summary, dialogue = research_paper_processor(file_path, selected_roles)

        return render_template('results.html', summary=summary, dialogue=dialogue)

@app.route('/download/<filename>')
def download(filename):
    return send_file(filename, as_attachment=True)

def research_paper_processor(pdf_path, selected_roles):
    """Main processing pipeline"""
    start_time = time.time()

    # Process paper
    text = extract_text(pdf_path)
    summary = generate_summary(text, llm)
    dialogue = generate_dialogue(text, selected_roles, llm)

    # Generate outputs
    create_audio(dialogue, selected_roles)
    create_summary_visual(summary)
    create_ppt(summary, dialogue)

    print(f"Total processing time: {time.time() - start_time:.2f} seconds")
    return summary, dialogue

# Run the Flask app
if __name__ == '__main__':
    app.run(host='0.0.0.0', port=5000)
```
